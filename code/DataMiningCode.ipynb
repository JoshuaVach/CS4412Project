```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

try:
    df = pd.read_csv('vgsales.csv')
    
    df = df.dropna(subset=['Year', 'Global_Sales', 'JP_Sales', 'NA_Sales', 'EU_Sales'])
    rows_to_drop = df[df['Year'] == 2020].index #only one in 2020 outlier
    df = df.drop(rows_to_drop)
    df = df[df['Global_Sales'] > 0] # Avoid division by zero
    
    df['JP_Percent'] = df['JP_Sales'] / df['Global_Sales']
    df['NA_Percent'] = df['NA_Sales'] / df['Global_Sales']
    df['EU_Percent'] = df['EU_Sales'] / df['Global_Sales']
    df['Year'] = df['Year'].astype(int)
    
    features = ['JP_Percent', 'NA_Percent', 'EU_Percent', 'Year']
    X = df[features]
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    #Calculate Inertia for different values of K 1-10
    inertia = []
    K_range = range(1, 11)
    
    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(X_scaled)
        inertia.append(kmeans.inertia_)
    
    plt.figure(figsize=(10, 6))
    plt.plot(K_range, inertia, marker='o', linestyle='--', color='teal')
    
    plt.title('The Elbow Method for Optimal K', fontsize=16)
    plt.xlabel('Number of Clusters (K)', fontsize=12)
    plt.ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
    plt.xticks(K_range)
    plt.grid(True, linestyle=':', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("Error: 'vgsales.csv' not found. Ensure the file is in your directory.")
except Exception as e:
    print(f"An error occurred: {e}")    
#picked k=4    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#Checking how each publisher does in each market depending on their total sales over time (k found using elbow method above) using to answer
# discovery question 1

try:
    df = pd.read_csv('vgsales.csv')

    # Data Cleaning: Remove missing values and games with 0 global sales
    df = df.dropna(subset=['Year', 'Publisher', 'Platform'])
    rows_to_drop = df[df['Year'] == 2020].index #only one in 2020 outlier
    df = df.drop(rows_to_drop)
    df = df[df['Global_Sales'] > 0]
    df['Year'] = df['Year'].astype(int)

    #Feature Engineering
    # calculate what % of total sales comes from each region
    # This standardizes the data
    df['JP_Percent'] = df['JP_Sales'] / df['Global_Sales']
    df['NA_Percent'] = df['NA_Sales'] / df['Global_Sales']
    df['EU_Percent'] = df['EU_Sales'] / df['Global_Sales']

    #Selecting Features for Clustering
    features = ['JP_Percent', 'NA_Percent', 'EU_Percent', 'Year']
    X = df[features]

    # Preprocessing: Scaling
    # K-Means calculates distance, so we must scale 'Year'
    # and 'Percentages' (0.0 to 1.0) to the same range.
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
    df['Cluster'] = kmeans.fit_predict(X_scaled)

    #Analysis
    cluster_summary = df.groupby('Cluster')[features].mean()
    print("--- Cluster Characteristics (Averages) ---")
    print(cluster_summary)

    #Visualization: Sales Profile by Cluster
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=df, x='Year', y='JP_Percent', hue='Cluster', palette='viridis', alpha=0.5)
    plt.title('Clusters of Game Success: Japan Market Share vs. Year', fontsize=15)
    plt.axhline(0.5, ls='--', color='red', alpha=0.3) # 50% threshold line
    plt.ylabel('Percentage of Sales from Japan')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=df, x='Year', y='NA_Percent', hue='Cluster', palette='viridis', alpha=0.5)
    plt.title('Clusters of Game Success: North America Market Share vs. Year', fontsize=15)
    plt.axhline(0.5, ls='--', color='red', alpha=0.3) # 50% threshold line
    plt.ylabel('Percentage of Sales from North America')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=df, x='Year', y='EU_Percent', hue='Cluster', palette='viridis', alpha=0.5)
    plt.title('Clusters of Game Success: Europe Market Share vs. Year', fontsize=15)
    plt.axhline(0.5, ls='--', color='red', alpha=0.3) # 50% threshold line
    plt.ylabel('Percentage of Sales from Europe')
    plt.show()

    print("\n--- Top Platforms per Cluster ---")
    for i in range(4):
        top_platforms = df[df['Cluster'] == i]['Platform'].value_counts().head(3).index.tolist()
        print(f"Cluster {i} Top Platforms: {top_platforms}")

except FileNotFoundError:
    print("Error: 'vgsales.csv' not found.")
except Exception as e:
    print(f"An error occurred: {e}")    
#interpretation: Cluster 2 is the most interesting as it is almost entirely made up of Japanese sales. All other clusters have at least 10%
# of other markets sales but japan seems to have a more robust domestic video game market that focuses less on foriegn markets. Another 
#interesting point is that cluster 0 and 1 both are very North American centric whole beingsplit over different time frames(0-2000 and 1-2008)
#Europe seems to have had a bigger push domestically in more recent years due to its higher percentage of sales occurs later.
#The older North american centric cluster90) had mainly PS PS2 and gameboy games, this shows that most likely these are japanese gaming companies
#selling well in the north american market while the later one (1) also has japanese consoles(wii and DS) it also has an american made xbox360
#this shows that while japanese companies still had a massive hold on the american market american corporations began moving in around this time
#The european cluster(3) is interesting because its largest console is pc followed by ps3 and xbox360 behind it. It shows a much more balanced
#market not dominated by any console and shows that games that do well in europe and no as well elsewhere typically are pc games.there is also 
#one game in 2020 which should not

  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

try:
    df = pd.read_csv('vgsales.csv')
    df = df.dropna(subset=['Year', 'Platform', 'Global_Sales'])
    rows_to_drop = df[df['Year'] == 2020].index #only one in 2020 outlier
    df = df.drop(rows_to_drop)
    df['Year'] = df['Year'].astype(int)

    # Total sales for each platform in each year
    platform_yearly = df.groupby(['Year', 'Platform'])['Global_Sales'].agg(['sum', 'max']).reset_index()
    platform_yearly.columns = ['Year', 'Platform', 'Global_Sales_Total', 'Global_Sales_Max']

    # Concentration Ratio = Sales of Top Game / Total Platform Sales
    platform_yearly['Concentration_Ratio'] = (
        platform_yearly['Global_Sales_Max'] / platform_yearly['Global_Sales_Total']
    )

    features = ['Global_Sales_Total', 'Concentration_Ratio']
    X = platform_yearly[features]


    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    
    inertia = []
    K_range = range(1, 11)  # Checking K from 1 to 10
    
    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(X_scaled)
        inertia.append(kmeans.inertia_)

    plt.figure(figsize=(10, 6))
    plt.plot(K_range, inertia, marker='o', linestyle='--', color='teal')
    
    plt.title('The Elbow Method for Optimal K', fontsize=16)
    plt.xlabel('Number of Clusters (K)', fontsize=12)
    plt.ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
    plt.xticks(K_range)
    plt.grid(True, linestyle=':', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("Error: 'vgsales.csv' not found. Please ensure the file is in your directory.")
except Exception as e:
    print(f"An error occurred: {e}")  
#k=3        
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#preparing to answer second discovery question using k means to determine the total sales of a platform per year and compare it to the 
# highest selling game that year to determine if one game helped to push sales high for a console

try:
    df = pd.read_csv('vgsales.csv')
    df = df.dropna(subset=['Year', 'Platform', 'Global_Sales'])
    df['Year'] = df['Year'].astype(int)


    # Total sales for each platform each year
    platform_yearly_sales = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().reset_index()

    # Find the sales of the #1 game for each platform each year
    top_game_sales = df.groupby(['Year', 'Platform'])['Global_Sales'].max().reset_index()

    # Merge these to calculate the "Concentration Ratio"
    platform_stats = pd.merge(platform_yearly_sales, top_game_sales, on=['Year', 'Platform'], suffixes=('_Total', '_Max'))
    platform_stats['Concentration_Ratio'] = platform_stats['Global_Sales_Max'] / platform_stats['Global_Sales_Total']


    #cluster based on Total Sales and the Concentration Ratio
    features = ['Global_Sales_Total', 'Concentration_Ratio']
    X = platform_stats[features]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Run K-Means 
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    platform_stats['Cluster'] = kmeans.fit_predict(X_scaled)

    # Visualization: Concentration vs Total Sales
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data=platform_stats, x='Global_Sales_Total', y='Concentration_Ratio', 
                    hue='Cluster', palette='Set1', alpha=0.6)

    plt.title('Platform Success Profiles: Total Sales vs. Game Concentration', fontsize=15)
    plt.xlabel('Total Platform Sales for the Year (Millions)')
    plt.ylabel('Concentration Ratio (Top Game Sales / Total Sales)')
    plt.axhline(0.5, ls='--', color='gray', alpha=0.5) # 50% marker
    

    plt.show()

    # Identifying the Dominant Platform each year and its Cluster
    idx = platform_stats.groupby(['Year'])['Global_Sales_Total'].transform(max) == platform_stats['Global_Sales_Total']
    yearly_leaders = platform_stats[idx].sort_values('Year')
    
    print("--- Yearly Platform Leaders and their Success Profile ---")
    print(yearly_leaders[['Year', 'Platform', 'Global_Sales_Total', 'Concentration_Ratio', 'Cluster']].tail(10))

except FileNotFoundError:
    print("Error: 'vgsales.csv' not found.")
except Exception as e:
    print(f"An error occurred: {e}")    

#interpretation: Cluster 0 is mainly poorly performing consoles that have one game that does well while cluster 1 is just poor to ok performing 
# consoles. Cluster 2 is where it gets interesting as most of them are well performing consoles that do not have one game dominating the market
# there is however an outlier that has over 150,000,000 sales with around 60% of them being from one game. The wii also suprised me as it has 
#highest selling game of all time but is not dominated by any one game. 
```

```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt

try:
    df = pd.read_csv('vgsales.csv').dropna()

    # Create  indicators for all three major regions
    df['High_JP'] = df['JP_Sales'] > df['JP_Sales'].median()
    df['High_NA'] = df['NA_Sales'] > df['NA_Sales'].median()
    df['High_EU'] = df['EU_Sales'] > df['EU_Sales'].median()

    
    def prepare_transactions(row):
    
        transaction = [
            f"Pub_{row['Publisher']}",
            f"Plat_{row['Platform']}",
            f"Era_{int(row['Year'] // 10 * 10)}s", # Grouping by decade
        ]
        # Append success tags if the game performed well in that region(data transformation)
        if row['High_JP']: transaction.append("Success_JP")
        if row['High_NA']: transaction.append("Success_NA")
        if row['High_EU']: transaction.append("Success_EU")
        return transaction

    dataset = df.apply(prepare_transactions, axis=1).tolist()

    #fp-growth
    te = TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df_fp = pd.DataFrame(te_ary, columns=te.columns_)

    frequent_itemsets = fpgrowth(df_fp, min_support=0.01, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.1)
    
    fig, axes = plt.subplots(1, 3, figsize=(20, 8), sharex=True)
    regions = [('Success_NA', 'North America', 'skyblue'), 
           ('Success_EU', 'Europe', 'orchid'), 
           ('Success_JP', 'Japan', 'salmon')]

    for i, (tag, title, color) in enumerate(regions):
        # Filter rules that result in success for this specific region
        reg_rules = rules[rules['consequents'].astype(str).str.contains(tag)]
    
        # Take the top 10 most "confident" rules
        top_reg = reg_rules.sort_values('confidence', ascending=False).head(10)
    
        # Create labels from the the conditions like Publisher and Platform
        labels = [", ".join(list(x)) for x in top_reg['antecedents']]
    
        axes[i].barh(labels, top_reg['confidence'], color=color)
        axes[i].set_title(f"Top Predictors for {title}", fontsize=16, fontweight='bold')
        axes[i].set_xlabel('Confidence (Probability of Success)', fontsize=12)
        axes[i].invert_yaxis()  # Best rules at the top
        axes[i].grid(axis='x', linestyle='--', alpha=0.5)

    plt.suptitle('Market Success DNA: Comparing Regional Predictors', fontsize=22, y=1.05)
    plt.tight_layout()
    plt.show()
    
except FileNotFoundError:
    print("Error: 'vgsales.csv' not found.")
except Exception as e:
    print(f"An error occurred: {e}")  
#This shows a lot of how success in video games rarely actually has anything to do with the publisher of the game in the 3 main markets what has 
# a much larger impact is the platform where the gamecube, playstation, gameboy advance, saturn, SNES and xbox all have a higher importance than
# the publisher. In fact only one publisher appears in the top predictors which is nintendo in Japan which shows that Nintendo does incredibly
# well inside of japan as well as North America enough to make games they make succeed at a higher rate. This also corroborates the findings in the previous k-means study that games sold 
# the most during the 90s 2000s and 2010s as all of these appear as predictors for all marketswith NA having mainly 2000s and europe and japan
# having more varied ones with all three or just 2010s and 1990s for japan with the 2000s and 1990s in all three markets being much highers than
# the 2010s as indicators. This shows that while the 2010s still had a good amount of success it was not quite as massive as the 90s and 2000s.
# This also shows that during the 90s in japan the saturn and SNES games did well while in North America in 2000s the game cube, xbox did incredibly
# well comparitively to other games.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

try:
    df = pd.read_csv('vgsales.csv').dropna(subset=['Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Global_Sales'])
    
    #use regional sales to find games that didn't follow the normal market split
    features = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
    X = df[features]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # contamination=0.001 means we want to flag the top .1% most unusual games
    iso_forest = IsolationForest(n_estimators=100, contamination=0.001, random_state=42)
    
    # Predict: -1 is an anomaly, 1 is normal
    df['Anomaly_Flag'] = iso_forest.fit_predict(X_scaled)
    # Get the raw score (lower = more anomalous)
    df['Anomaly_Score'] = iso_forest.decision_function(X_scaled)

    anomalies = df[df['Anomaly_Flag'] == -1].sort_values('Anomaly_Score')

    print(f"Detected {len(anomalies)} anomalies.")
    print("\n--- Top 10 Most Anomalous Games ---")
    print(anomalies[['Name', 'Platform', 'Year', 'Publisher', 'Global_Sales']].head(10))

    #NA graph
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data=df, x='NA_Sales', y='Global_Sales', 
                    hue='Anomaly_Flag', palette={1: 'blue', -1: 'red'}, 
                    alpha=0.6, s=50)
    
    plt.title('Anomaly Detection: Regional Outliers', fontsize=15)
    plt.xlabel('North American Sales')
    plt.ylabel('Global Sales')
    plt.legend(title='Status', labels=['Normal', 'Anomaly'])
    plt.show()

    #Japan Graph
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data=df, x='JP_Sales', y='Global_Sales', 
                    hue='Anomaly_Flag', palette={1: 'blue', -1: 'red'}, 
                    alpha=0.6, s=50)
    
    plt.title('Anomaly Detection: Regional Outliers', fontsize=15)
    plt.xlabel('Japan Sales')
    plt.ylabel('Global Sales')
    plt.legend(title='Status', labels=['Normal', 'Anomaly'])
    plt.show()

    #EU graph
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data=df, x='EU_Sales', y='Global_Sales', 
                    hue='Anomaly_Flag', palette={1: 'blue', -1: 'red'}, 
                    alpha=0.6, s=50)
    
    plt.title('Anomaly Detection: Regional Outliers', fontsize=15)
    plt.xlabel('European Sales')
    plt.ylabel('Global Sales')
    plt.legend(title='Status', labels=['Normal', 'Anomaly'])
    plt.show()

except FileNotFoundError:
    print("Error: 'vgsales.csv' not found.")
except Exception as e:
    print(f"An error occurred: {e}")     

#this detection has found a few interesting things although most are expected like wii sports and a bunch of popular wii and ds games which 
# are the most popular games of all time. The one i was least expecting was the Brain age whichi have never heard of until this test which apparently 
# has half of its total sales in europe which is pretty anomalous the other anomalies also tend to be clustered the most in the EU chart 
#which goes to show that selling well in europe as well as other markets is rare enough to warrant being marked as anomalous.
```
